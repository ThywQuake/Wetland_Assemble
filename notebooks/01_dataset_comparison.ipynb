# 01_dataset_comparison.ipynb

本项目用于多源湿地数据集的局部空间对比（GWD30 vs GLWD vs GIEMS-MC）。

```python
import sys, os
sys.path.append("../src")
import xarray as xr
import geemap
import ipyleaflet
from wetland_analysis.data.loader import load_wetland_dataset, list_available_datasets

# 远程服务器代理配置 (Required for remote visualization)
os.environ['LOCALTILESERVER_CLIENT_PREFIX'] = 'proxy/{port}'
```

## 1. 区域与数据加载

```python
datasets = list_available_datasets()
print(f"Available datasets: {datasets}")

# 示例：加载 2020 年亚马逊区域的 GWD30
gwd30 = load_wetland_dataset('gwd30', year=2020)
print("GWD30 Loaded.")
```

## 2. 交互式对比 (SplitPanel)

您可以左右拖动滑块查看不同分辨率数据集的差异。

```python
m = geemap.Map(center=[-3, -60], zoom=10) # 亚马逊样区

left_layer = geemap.xarray_tile(gwd30, colormap='Blues', name='GWD30 2020')
# 假设另一个加载好的低分数据
right_layer = geemap.basemap_tile('SATELLITE')

control = ipyleaflet.SplitControl(left_layer=left_layer, right_layer=right_layer)
m.add_control(control)
m
```
